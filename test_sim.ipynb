```python
%%time
surpluses = []

for _ in tqdm(range(10000)):
    sim = SimulatorSampledArrival(num_background_agents=25, 
                                  sim_time=12000, 
                                  lam=5e-4, 
                                  mean=1e5, 
                                  r=0.05, 
                                  shock_var=5e6, 
                                  q_max=10,
                                  pv_var=5e6,
                                  shade=[250,500])
    sim.run()
    fundamental_val = sim.markets[0].get_final_fundamental()
    values = []
    for agent_id in sim.agents:
        agent = sim.agents[agent_id]
        value = agent.get_pos_value() + agent.position * fundamental_val + agent.cash
        # print(agent.cash, agent.position, agent.get_pos_value(), value)
        values.append(value)
    surpluses.append(sum(values)/len(values))
print(sum(surpluses)/len(surpluses)*25)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

market = sim.markets[0]

# -------------------------
# 1. Build trade dataframe
# -------------------------
trade_rows = []
for mo in market.matched_orders:
    trade_rows.append({
        "exec_time": mo.time,
        "exec_price": mo.price,
        "order_time": mo.order.time,
        "order_type": mo.order.order_type,   # 1 buy, -1 sell
        "quantity": mo.order.quantity,
        "agent_id": mo.order.agent_id,
        "order_id": mo.order.order_id,
        "asset_id": mo.order.asset_id,
        "order_price": mo.order.price
    })

trades = pd.DataFrame(trade_rows)
print("Raw matched orders:", len(trades))
display(trades.head())

# -------------------------
# 2. Deduplicate executions
# -------------------------
# Each execution seems to appear twice (once per side).
# For tonight, deduplicate by execution time + execution price + asset.
# If quantity varies later, include quantity too.
trades_unique = (
    trades
    .drop_duplicates(subset=["exec_time", "exec_price", "asset_id"])
    .sort_values("exec_time")
    .reset_index(drop=True)
)

print("Unique executions:", len(trades_unique))
display(trades_unique.head())

# -------------------------
# 3. Price series and returns
# -------------------------
# Option A: use executed trade prices
trade_prices = trades_unique[["exec_time", "exec_price"]].copy()
trade_prices["return"] = np.log(trade_prices["exec_price"]).diff()

# Option B: use midprices if available
mid_df = pd.DataFrame({
    "t": np.arange(len(market.order_book.midprices)),
    "midprice": market.order_book.midprices
})
mid_df = mid_df.dropna().copy()
mid_df["return"] = np.log(mid_df["midprice"]).diff()

print("Trade-price observations:", len(trade_prices))
print("Midprice observations:", len(mid_df))

# Pick one main series for comparison.
# I recommend midprice if it is sufficiently populated; otherwise trade prices.
if len(mid_df.dropna()) > 20:
    analysis_df = mid_df.rename(columns={"t": "time", "midprice": "price"})
    price_col = "price"
    series_name = "midprice"
else:
    analysis_df = trade_prices.rename(columns={"exec_time": "time", "exec_price": "price"})
    price_col = "price"
    series_name = "trade price"

analysis_df = analysis_df.dropna().reset_index(drop=True)
returns = analysis_df["return"].dropna()

print(f"Using {series_name} series for analysis.")
print("Return observations:", len(returns))

# -------------------------
# 4. Heavy tails
# -------------------------
kurt = returns.kurtosis()
print("Kurtosis:", kurt)

# -------------------------
# 5. Volatility clustering
# -------------------------
vol_clust = (returns ** 2).autocorr(lag=1)
print("Volatility clustering (lag-1 autocorr of squared returns):", vol_clust)

# -------------------------
# 6. Aggregational Gaussianity
# -------------------------
# Compare returns aggregated over larger horizons
def aggregated_log_returns(price_series, horizons=(1, 5, 10)):
    out = {}
    logp = np.log(price_series.dropna())
    for h in horizons:
        out[h] = logp.diff(h).dropna()
    return out

agg_returns = aggregated_log_returns(analysis_df[price_col], horizons=(1, 5, 10))

for h, r in agg_returns.items():
    print(f"Horizon {h}: n={len(r)}, mean={r.mean():.6f}, std={r.std():.6f}, kurtosis={r.kurtosis():.6f}")

# -------------------------
# 7. Plots
# -------------------------
import os
save_dir = r"R:\sescott1\Masters\thesis\potential_codes\pymarketsim\sarah\26_march"
os.makedirs(save_dir, exist_ok=True)

plt.figure(figsize=(8, 4))
plt.plot(analysis_df["time"], analysis_df[price_col])
plt.title(f"{series_name.title()} Series")
plt.xlabel("Time")
plt.ylabel("Price")
plt.savefig(os.path.join(save_dir, f"{series_name.title()}_series.png"), dpi=300, bbox_inches="tight")
plt.show()

# Return Distribution
plt.figure(figsize=(8, 4))
plt.hist(returns, bins=30)
plt.title(f"Return Distribution ({series_name})")
plt.xlabel("Return")
plt.ylabel("Frequency")
plt.savefig(os.path.join(save_dir, f"return_distribution_{series_name}.png"), dpi=300, bbox_inches="tight")
plt.show()

# Squared returns for volatility clustering
plt.figure(figsize=(8, 4))
plt.plot((returns ** 2).reset_index(drop=True))
plt.title("Squared Returns")
plt.xlabel("Index")
plt.ylabel("Squared Return")
plt.savefig(os.path.join(save_dir, "squared_returns.png"), dpi=300, bbox_inches="tight")
plt.show()

# Optional: histograms across horizons for aggregational Gaussianity
for h, r in agg_returns.items():
    plt.figure(figsize=(8, 4))
    plt.hist(r, bins=30)
    plt.title(f"Aggregated Returns (h={h})")
    plt.xlabel("Return")
    plt.ylabel("Frequency")
    plt.savefig(os.path.join(save_dir, f"aggregated_returns_h={h}.png"), dpi=300, bbox_inches="tight")
    plt.show()

# -------------------------
# 8. Final spread snapshot only
# -------------------------
final_best_bid = market.order_book.get_best_bid()
final_best_ask = market.order_book.get_best_ask()
final_spread = final_best_ask - final_best_bid

print("Final best bid:", final_best_bid)
print("Final best ask:", final_best_ask)
print("Final spread snapshot:", final_spread)
```

```python
# Normalized returns
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

market = sim.markets[0]

# -------------------------
# 1. Build trade dataframe
# -------------------------
trade_rows = []
for mo in market.matched_orders:
    trade_rows.append({
        "exec_time": mo.time,
        "exec_price": mo.price,
        "order_time": mo.order.time,
        "order_type": mo.order.order_type,
        "quantity": mo.order.quantity,
        "agent_id": mo.order.agent_id,
        "order_id": mo.order.order_id,
        "asset_id": mo.order.asset_id,
        "order_price": mo.order.price
    })

trades = pd.DataFrame(trade_rows)
print("Raw matched orders:", len(trades))
display(trades.head())

# -------------------------
# 2. Deduplicate executions
# -------------------------
trades_unique = (
    trades
    .drop_duplicates(subset=["exec_time", "exec_price", "asset_id"])
    .sort_values("exec_time")
    .reset_index(drop=True)
)

print("Unique executions:", len(trades_unique))
display(trades_unique.head())

# -------------------------
# 3. Price series and returns
# -------------------------
trade_prices = trades_unique[["exec_time", "exec_price"]].copy()
trade_prices["return"] = np.log(trade_prices["exec_price"]).diff()

mid_df = pd.DataFrame({
    "t": np.arange(len(market.order_book.midprices)),
    "midprice": market.order_book.midprices
})
mid_df = mid_df.dropna().copy()
mid_df["return"] = np.log(mid_df["midprice"]).diff()

print("Trade-price observations:", len(trade_prices))
print("Midprice observations:", len(mid_df))

if len(mid_df.dropna()) > 20:
    analysis_df = mid_df.rename(columns={"t": "time", "midprice": "price"})
    price_col = "price"
    series_name = "midprice"
else:
    analysis_df = trade_prices.rename(columns={"exec_time": "time", "exec_price": "price"})
    price_col = "price"
    series_name = "trade price"

analysis_df = analysis_df.dropna().reset_index(drop=True)
returns = analysis_df["return"].dropna().astype(float)

print(f"Using {series_name} series for analysis.")
print("Return observations:", len(returns))

# -------------------------
# 4. Raw metrics
# -------------------------
kurt = returns.kurtosis()
print("Kurtosis:", kurt)

vol_clust = (returns ** 2).autocorr(lag=1)
print("Volatility clustering (lag-1 autocorr of squared returns):", vol_clust)

# -------------------------
# 5. Aggregational Gaussianity
# -------------------------
def aggregated_log_returns(price_series, horizons=(1, 5, 10)):
    out = {}
    logp = np.log(price_series.dropna())
    for h in horizons:
        out[h] = logp.diff(h).dropna()
    return out

agg_returns = aggregated_log_returns(analysis_df[price_col], horizons=(1, 5, 10))

for h, r in agg_returns.items():
    print(
        f"Horizon {h}: n={len(r)}, mean={r.mean():.6f}, std={r.std():.6f}, kurtosis={r.kurtosis():.6f}"
    )

# -------------------------
# 6. Standardize for visual comparison
# -------------------------
z_returns = (returns - returns.mean()) / returns.std()

z_agg_returns = {}
for h, r in agg_returns.items():
    z_agg_returns[h] = (r - r.mean()) / r.std()

# -------------------------
# 7. Save location
# -------------------------
save_dir = r"R:\sescott1\Masters\thesis\potential_codes\pymarketsim\sarah\26_march"
os.makedirs(save_dir, exist_ok=True)

# -------------------------
# 8. Price series
# -------------------------
plt.figure(figsize=(8, 4))
plt.plot(analysis_df["time"], analysis_df[price_col])
plt.title(f"Raw {series_name.title()} Series")
plt.xlabel("Time")
plt.ylabel("Price")
plt.savefig(os.path.join(save_dir, f"{series_name}_series_raw.png"), dpi=300, bbox_inches="tight")
plt.show()

# -------------------------
# 9. Raw return distribution
# -------------------------
plt.figure(figsize=(8, 4))
plt.hist(returns, bins=30, density=True)
plt.title(f"Raw Return Distribution ({series_name})")
plt.xlabel("Return")
plt.ylabel("Density")
plt.savefig(os.path.join(save_dir, f"return_distribution_{series_name}_raw.png"), dpi=300, bbox_inches="tight")
plt.show()

# -------------------------
# 10. Standardized return distribution
# -------------------------
plt.figure(figsize=(8, 4))
plt.hist(z_returns, bins=30, density=True)
plt.title(f"Standardized Return Distribution ({series_name})")
plt.xlabel("Standardized return")
plt.ylabel("Density")
plt.savefig(os.path.join(save_dir, f"return_distribution_{series_name}_standardized.png"), dpi=300, bbox_inches="tight")
plt.show()

# -------------------------
# 11. Squared returns
# -------------------------
plt.figure(figsize=(8, 4))
plt.plot((returns ** 2).reset_index(drop=True))
plt.title(f"Raw Squared Returns ({series_name})")
plt.xlabel("Index")
plt.ylabel("Squared Return")
plt.savefig(os.path.join(save_dir, f"squared_returns_{series_name}_raw.png"), dpi=300, bbox_inches="tight")
plt.show()

# -------------------------
# 12. Aggregated standardized return histograms
# -------------------------
for h, r in z_agg_returns.items():
    plt.figure(figsize=(8, 4))
    plt.hist(r, bins=30, density=True)
    plt.title(f"Standardized Aggregated Returns (h={h})")
    plt.xlabel("Standardized return")
    plt.ylabel("Density")
    plt.savefig(os.path.join(save_dir, f"aggregated_returns_h={h}_standardized.png"), dpi=300, bbox_inches="tight")
    plt.show()

# -------------------------
# 13. Final spread snapshot
# -------------------------
final_best_bid = market.order_book.get_best_bid()
final_best_ask = market.order_book.get_best_ask()
final_spread = final_best_ask - final_best_bid

print("Final best bid:", final_best_bid)
print("Final best ask:", final_best_ask)
print("Final spread snapshot:", final_spread)
```
